```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread("scenery.jpg")
dst = cv2.imread("plane.jpg")
src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(src)
```

```python
plt.imshow(dst)
```

```python
mask = 255 * np.ones(dst.shape, dst.dtype)
```

```python
width, height, channels = src.shape
center = (int(height/2),int(width/2))
center = (800,100)
```

```python
normal_clone = cv2.seamlessClone(dst,src, mask, center, cv2.NORMAL_CLONE)
mixed_clone = cv2.seamlessClone(dst,src, mask, center, cv2.MIXED_CLONE)
```

```python
plt.imshow(normal_clone,interpolation="none")
```

```python
plt.imshow(mixed_clone)
```
